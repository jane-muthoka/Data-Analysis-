```python
import pandas as pd
import numpy as numpy
import seaborn as sns
import matplotlib as plt
import os
```

```python
alcodf=pd.read_csv('cardio_alco.csv', sep=';')
basedf=pd.read_csv('cardio_base.csv', sep=',')
covidf=pd.read_csv('covid_data.csv', sep=',')
covidf.head()
```

```python
basedf.head()
```

```python
#convert age from days to years
basedf['age']=basedf['age']//365
```

```python
# return row(s) with the maximum age
max_age_rows = basedf[basedf['age'] == basedf['age'].max()]
max_age_rows
```

```python
# average weight for each age (age already converted to years in basedf)
avg_weight_by_age = basedf.groupby('age', as_index=False)['weight'].mean().rename(columns={'weight': 'avg_weight'})
avg_weight_by_age['avg_weight'] = avg_weight_by_age['avg_weight'].round(2)
avg_weight_by_age = avg_weight_by_age.sort_values('avg_weight', ascending=False)
avg_weight_by_age
```

```python
#how higher is people over 50 holestrol levels than the rest
over_50 = basedf[basedf['age'] > 50]['cholesterol'].mean()
under_50 = basedf[basedf['age'] <= 50]['cholesterol'].mean()
difference =(over_50 - under_50) *100 / under_50
difference
```

```python
#are men likely to be a smoker than women
men_smoker = basedf[(basedf['gender'] == 1) & (basedf['smoke'] == 1)]
women_smoker = basedf[(basedf['gender'] == 0) & (basedf['smoke'] == 1)]
men_smoker_count = men_smoker.shape[0]
women_smoker_count = women_smoker.shape[0]
men_smoker_count, women_smoker_count
```
